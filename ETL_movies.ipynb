```python
# Transformo cada fila con elementos de tipo "str" a "list" para acceder a cada valor individualmente y se lo paso a una variable para no dañar
# nada del df original.
belongs_to_collection_separada = []
    
# Se recorre todas las filas de la columna "belongs_to_collection"
for i in range(len(df_movies["belongs_to_collection"].index)):
    if type(df_movies["belongs_to_collection"][i]) == str:
        belongs_to_collection_separada.append(df_movies["belongs_to_collection"][i].split(","))
    else:
        belongs_to_collection_separada.append(df_movies["belongs_to_collection"][i])
```

Solo se requiere el valor de cada par de elementos, por lo que procederé a eliminar las partes "claves".

```python
# Se crean listas donde serán almacenados los valores de cada "clave" respectivamente
id_belongs_to_collection = []
name_belongs_to_collection = []
poster_path_belongs_to_collection = []
backdrop_path_belongs_to_collection = []

# Se recorre el completo de la variable (todas sus "filas")
for i in range(len(belongs_to_collection_separada)):
    # El total de claves son 4, me aseguro que sea tipo lista y solo por si acaso, que hayan más "claves" que 0 y menos que 5
    if type(belongs_to_collection_separada[i]) == list and len(belongs_to_collection_separada[i]) > 0 and len(belongs_to_collection_separada[i]) < 5:
        if len(belongs_to_collection_separada[i]) == 1:
            id_belongs_to_collection.append(belongs_to_collection_separada[i][0].replace("{'id': ", "").strip())
            name_belongs_to_collection.append("nan")
            poster_path_belongs_to_collection.append("nan")
            backdrop_path_belongs_to_collection.append("nan")

        elif len(belongs_to_collection_separada[i]) == 2:
            id_belongs_to_collection.append(belongs_to_collection_separada[i][0].replace("{'id': ", "").strip())
            name_belongs_to_collection.append(belongs_to_collection_separada[i][1].replace(" 'name': ", "").strip())
            poster_path_belongs_to_collection.append("nan")
            backdrop_path_belongs_to_collection.append("nan")

        elif len(belongs_to_collection_separada[i]) == 3:
            id_belongs_to_collection.append(belongs_to_collection_separada[i][0].replace("{'id': ", "").strip())
            name_belongs_to_collection.append(belongs_to_collection_separada[i][1].replace(" 'name': ", "").strip())
            poster_path_belongs_to_collection.append(belongs_to_collection_separada[i][2].replace(" 'poster_path': ", "").strip())
            backdrop_path_belongs_to_collection.append("nan")

        elif len(belongs_to_collection_separada[i]) == 4:
            id_belongs_to_collection.append(belongs_to_collection_separada[i][0].replace("{'id': ", "").strip())
            name_belongs_to_collection.append(belongs_to_collection_separada[i][1].replace(" 'name': ", "").strip())
            poster_path_belongs_to_collection.append(belongs_to_collection_separada[i][2].replace(" 'poster_path': ", "").strip())
            backdrop_path_belongs_to_collection.append(belongs_to_collection_separada[i][2].replace(" 'backdrop_path': ", "").strip())

    else:
        id_belongs_to_collection.append("nan")
        name_belongs_to_collection.append("nan")
        poster_path_belongs_to_collection.append("nan")
        backdrop_path_belongs_to_collection.append("nan")
```

Función que se encarga de obtener los retornos

```python
def retorno_revenue_budget():
    """Esta función se encarga de retornar el retorno de ganancia por cada fila entre revenue y budget, dividiendo revenue entre budget"""
    retorno = []
    #Accedo a todos los índices con un bucle
    for i in df_movies.index:
        #Compruebo que ambas filas tengas valores mayores a 0
        if df_movies["revenue"][i] > 0 and df_movies["budget"][i] > 0:
            retorno.append(df_movies["revenue"][i] / df_movies["budget"][i])
        #En otro caso, se agregará 0
        else:
            retorno.append(0)
    
    return retorno

retorno = retorno_revenue_budget()
```
